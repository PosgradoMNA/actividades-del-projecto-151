Fundamentos de bases de datos y para ciencia de datos.

Las bases de datos para ciencia de datos son un elemento clave, ya que son gran parte de la materia prima para poder generar modelos analíticos. Las bases de datos pueden ser SQL(relacional) o NoSql dependiendo los herramientas con las que sean explotados los datos y con la estrategia de almacenamiento.


Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

Es un conjunto muy grande de datos, de diversas fuentes que se centraliza en un solo lugar. Comúnmente esta gran colección de datos permite tener un registro histórico de eventos. Constantemente un data warehouse esta ligado a un ETL para poder transportar y transformar los datos de origen hacia el data warehouse, un modelo relacional (si es el caso), una capa de análisis, clientes para que distintos usuarios puedan acceder a la información, entre otros.
Algunos conceptos clave dentro de un data warehouse son consistencia de información a lo largo de las fuentes, estabilidad, variante en el tiempo y orientada a un tema. Todos estos conceptos permiten que las datos puedan ser confiables y puedan convertirse en información una vez tratados los datos.
